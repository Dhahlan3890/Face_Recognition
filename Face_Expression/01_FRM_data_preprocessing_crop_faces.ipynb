```python
import numpy as np # numerical python
import cv2 # opencv
from glob import glob # glob use to extract path of file
import matplotlib.pyplot as plt # visualze and display 
```

```python
happypath = glob('./data/happy/*.jpg') # this will return all images path in a list
sadpath = glob('./data/sad/*.jpg')
angrypath = glob('./data/angry/*.jpg')
```

```python
print('The number of images in happy folder = ',len(happypath))
print('The number of images in sad folder = ',len(sadpath))
print('The number of images in angry folder = ',len(angrypath))
```

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
for i in range(len(happypath)):
    try:
            
        ### Step -1 Read Image and Convert to RGB
        img = cv2.imread(happypath[i]) # read image in BGR
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # this step will convert image from BGR to RGB

        ### Step -2: Apply Haar Cascade Classifier
        gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces_list:
            ### Step -3: crop Face 
            roi = img[y:y+h,x:x+w]
            ### Step -4: Save Image
            cv2.imwrite(f'./crop_data/happy/happy_{i}.jpg',roi)
            print('Image Sucessfully processed')
    except:
        print('Unable to Process the image')
```
